### Projeto
**Modelos de clusterização e classificação** para segmentação de clientes em campanhas de marketing em uma plataforma de delivery.

### Objetivo
O objetivo do projeto é, a partir dos dados de campanhas piloto, construir um **modelo de clusterização** para segmentação de clientes em campanhas de marketing em uma plataforma de delivery. <br>

A intenção é que o modelo preveja o comportamento dos consumidores e torne possível aplicá-lo a toda a base de clientes, permitindo que a empresa identifique os clientes que têm maior probabilidade de aceitar a campanha, otimizando assim os resultados das campanhas e tornando-as altamente lucrativas. <br>

E em seguida, construir um **modelo de classificação** de acordo com os resultados obtidos no modelo inicial para prever se os clientes devem ou não aceitar as futuras campanhas da empresa.

### Descrição

Nesse projeto de Ciência de Dados, resumidamente, foi realizada uma ampla análise exploratória gerando os principais insights do projeto e permitindo identificar duas principais personas para os clientes de uma plataforma de delivery. 

Em seguida, foi criado um modelo de clusterização usando K-Means para identificar melhor as personas dos clientes e permitindo assim a empresa direcionar melhor a campanha de marketing para os clientes que tem a maior probabilidade de aceitá-la. 

Por fim, foi criado um modelo de classificação usando Regressão Logística para prever se os clientes devem ou não aceitar as futuras campanhas de marketing da empresa.

### Pipeline do projeto
- **01. Pré Processamento - Pre Processing:** É a etapa onde os dados são preparados para serem usados nos modelos. As técnicas usadas são: One Hot Encoder, Standard Scaler, Power Transformation, Min-Max Scaler, Robust Scaler
- **02. Seleção de Colunas - Feature Selection:** É a etapa onde se determina as colunas mais relevantes para os modelos. As técnicas usadas são: SelectKBest, VarianceThreShold, RFE - Recursive Feature Elimination
- **03. Redução de Dimensionalidade - Dimensionality Reduction:** É a etapa onde se reduz a quantidade de features para simplificação dos modelos. As técnicas usadas são: PCA, LDA, t-SNE e UMAP
- **04. Reamostragem - Resampling:** É a etapa onde se equipara as classes de um dataset desbalanceado. As técnicas usadas são: RUS - Random Under Sampling, ROS - Random Over Sampling
- **05. Modelos - Models:** É a etapa onde se treina os modelos. As técnicas usadas são: Árvore de Classificação, Regressão Logística, KNN, Regressão Linear, Regressão Polinomial, Árvore de Regressão, K-Means, Mean-Shift, DBSCAN
- **06. Validação Cruzada - Cross Validation:** É a etapa onde divide os dados em várias partes para a avaliação dos modelos. Cada divisão feita do dataset é usada uma vez como teste e as demais vezes como treino. As técnicas usadas são: K-Fold, Stratified K-Fold
- **07. Otimização de Hiperparâmetros - Hyperparameter Tuning:** É a etapa onde se busca otimizar os parâmetros do modelo selecionado para que com uma performance ainda melhor. As técnicas usadas são: Grid Search, Random Search, Bayesian Optimization
- **08. Estimador Final - Final Estimator:** É a etapa onde se treina o modelo com os parâmetros otimizados
- **09. Resultados - Results:** É a etapa onde se avalia os resultados finais após o treinamento e otimização do modelo. As técnicas usadas são: Accuracy, Precision, Recall, F1-Score, ROC-AUC, AUPRC, AP, Confusion Matrix, Time, MAE, MSE, RMSE, R², Time, Elbow Method, Silhouette Method
