### Boxplot - Gráfico de caixa

```python
display(Image(consts.BOXPLOT, width=500)) # Exibindo o gráfico
```

O **boxplot (gráfico de caixa)** é uma ferramenta gráfica utilizada para exibir a distribuição de um conjunto de dados com base em quartis. Ele mostra a mediana, o intervalo interquartil (IQR), e possíveis valores atípicos. A caixa representa o IQR, enquanto as linhas (whiskers) indicam a extensão dos dados fora dos quartis. Pontos fora dessas linhas são considerados outliers. O boxplot é útil para identificar assimetria, dispersão e outliers em distribuições, especialmente em comparações entre múltiplos grupos.

### Scatterplot - Gráfico de dispersão

```python
display(Image(consts.SCATTERPLOT, width=500)) # Exibindo o gráfico
```

O **scatterplot (gráfico de dispersão)** é uma representação visual que exibe a relação entre duas variáveis numéricas. Cada ponto no gráfico corresponde a uma observação, onde a posição no eixo x representa o valor de uma variável e no eixo y, o valor da outra. Ele é amplamente utilizado para identificar correlações, padrões, tendências ou clusters nos dados, bem como para detectar possíveis outliers. Quanto mais concentrados os pontos em uma linha ou curva, maior a correlação entre as variáveis.

### Histplot - Histograma

```python
display(Image(consts.HISTPLOT, width=500)) # Exibindo o gráfico
```

O **histplot (histograma)** é uma representação gráfica que exibe a distribuição de um conjunto de dados numéricos. Ele é composto por barras, onde cada barra representa uma faixa (ou intervalo) de valores, chamada de bin, e sua altura indica a frequência (ou contagem) de observações dentro desse intervalo. O eixo horizontal (x) mostra os valores ou categorias, enquanto o eixo vertical (y) exibe a frequência. O histograma é útil para entender a forma da distribuição dos dados, identificando padrões como simetria, caudas longas, e a presença de picos ou lacunas. É frequentemente utilizado para visualizar a dispersão de dados contínuos. 

Para complementar o histograma, pode-se adicionar uma KDE (Kernel Density Estimate), que é uma curva suave que aproxima a distribuição dos dados. A KDE fornece uma visualização contínua da densidade dos dados, ajudando a identificar padrões e tendências que podem não ser tão evidentes no histograma sozinho. Essa combinação facilita a análise da forma da distribuição e a presença de múltiplos picos ou irregularidades.

### KDEplot - Gráfico de KDE

```python
display(Image(consts.KDEPLOT, width=500)) # Exibindo o gráfico
```

O **KDEplot (Kernel Density Estimate plot)** é uma visualização gráfica que estima a densidade de probabilidade de uma variável contínua. Ele gera uma curva suave que aproxima a distribuição dos dados, sem depender de intervalos definidos como no histograma. É útil para visualizar a distribuição subjacente de dados, identificar múltiplos picos (distribuições multimodais) e compreender a variação e dispersão dos dados. O KDEplot oferece uma visão mais refinada da distribuição em comparação com gráficos discretos, como o histograma.

### Heatmap - Mapa de calor

```python
display(Image(consts.HEATMAP, width=500)) # Exibindo o gráfico
```

O **heatmap (mapa de calor)** é uma representação gráfica que utiliza uma grade de cores para mostrar a magnitude dos valores em uma matriz. Cada célula da matriz é preenchida com uma cor que corresponde a um valor numérico, facilitando a identificação de padrões, correlações e anomalias. É comumente usado em análise de correlações, onde pode-se visualizar a força da relação entre variáveis, ou em contextos que envolvem grandes quantidades de dados organizados em tabelas.

### Pairplots - Gráficos de dispersão em pares

```python
display(Image(consts.PAIRPLOTS, width=500)) # Exibindo o gráfico
```

O **pairplot (gráfico de pares)** é uma ferramenta usada para visualizar as relações entre múltiplas variáveis de um dataset. Ele cria gráficos de dispersão para cada par de variáveis e gráficos de densidade ou histogramas nas diagonais para exibir a distribuição de cada uma individualmente. O pairplot é particularmente útil para identificar correlações entre variáveis, distribuições e possíveis interações entre dados, proporcionando uma visão geral abrangente do comportamento das variáveis numéricas no conjunto de dados.
