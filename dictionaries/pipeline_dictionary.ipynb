### Pre Processing (Pré Processamento)

**Objetivo:** Preparar os dados brutos para modelagem.

**Técnicas:** 
- One Hot Enconder
- Standard Scaler
- Power Transformation
- Min-Max Scaler
- Robust Scaler

**Importância:** Essencial para garantir que os dados estejam limpos e preparados para os modelos, evitando problemas de desempenho e qualidade.

### Feature Selection (Seleção de Colunas)

**Objetivo:** Eliminar variáveis irrelevantes ou redundantes que podem prejudicar o modelo.

**Técnicas:**
- SelectKBest
- VarianceThreShold
- RFE - Recursive Feature Elimination

**Importância:** Reduz a complexidade do modelo e melhora a interpretabilidade e performance ao focar nas variáveis mais relevantes

### Dimensionality Reduction (Redução de Dimensionalidade)

**Objetivo:** Reduzir o número de variáveis mantendo a maior parte da informação relevante.

**Técnicas:**
- PCA
- LDA
- t-SNE e UMAP

**Importância:** Reduz o risco de overfitting e melhora a eficiência computacional em conjuntos de dados com muitas variáveis.

### Resampling (Reamostragem)

**Objetivo:** Corrigir desequilíbrios entre classes no dataset, especialmente em problemas de classificação.

**Técnicas:**
- RUS (Random Under Sampling)
- ROS (Random Over Sampling)

**Importância:** Evita que os modelos fiquem enviesados em favor das classes majoritárias.

### Models (Modelos)

**Objetivo:** Aplicar diferentes algoritmos de machine learning e selecionar o mais adequado para o problema.

**Técnicas:**
- Classificação: Árvore de Classificação, Regressão Logística, KNN
- Regressão: Regressão Linear, Regressão Polinomial, Árvore de Regressão
- Clusterização: K-Means, Mean-Shift, DBSCAN

**Importância:** Escolher o modelo correto para o problema, considerando a natureza dos dados e a tarefa (classificação, regressão, etc.).

### Cross Validation (Validação Cruzada)

**Objetivo:** Avaliar a capacidade de generalização do modelo, garantindo que ele funcione bem em diferentes subconjuntos de dados.

**Técnicas:**
- K-Fold 
- Stratified K-Fold (Indicado para datasets desbalanceados, por manter a proporção das classes)

**Importância:** Reduz o risco de overfitting e fornece uma avaliação mais robusta do desempenho do modelo em diferentes partições dos dados.

### Hyperparameter Tuning (Otimização de Hiperparâmetros)

**Objetivo:** Melhorar a performance do modelo ajustando hiperparâmetros não aprendidos durante o treinamento (ex: taxa de aprendizado, número de árvores em Random Forest).

**Técnicas:**
- Grid Search
- Random Search
- Bayesian Optimization

**Importância:** Garantir que o modelo esteja ajustado para seu melhor desempenho possível, evitando sub ou sobreajuste.

### Final Estimator (Estimador Final)

**Objetivo:** Com os melhores hiperparâmetros encontrados, treinar o modelo final para uso em produção.

**Importância:** O modelo final será o responsável por realizar as previsões em novos dados, então precisa estar devidamente ajustado e otimizado.

### Results (Resultados)

**Objetivo:** Avaliar e comunicar a performance final do modelo.

**Técnicas:**
- Métricas Classificação: Accuracy, Precision, Recall, F1-Score, ROC-AUC, AUPRC, AP, Confusion Matrix, Time
- Métricas Regressão: MAE, MSE, RMSE, R², Time
- Métricas Clusterização: Elbow Method, Silhouette Method

**Importância:** Garantir que os resultados sejam interpretáveis e possam ser apresentados de maneira clara aos stakeholders. Esta etapa também inclui insights para decisões de negócio.
